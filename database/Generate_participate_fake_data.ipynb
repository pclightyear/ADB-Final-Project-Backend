```python
manage_dict = {}
for i in range(1000):
    manage_dict[str(i)] = manage[manage['UID'] == i].loc[:,'PID'].tolist()
```

```python
manage_dict
```

```python
import random

project_list = []
user_list = []

participate_num = 0

while True:
    uid = random.randint(0,999)
    pid = random.randint(0,999)
    if pid not in manage_dict[str(uid)]:
        user_list.append(uid)
        project_list.append(pid)
        participate_num+=1
    if participate_num == 10000:
        break
```

```python
project_list
```

```python
user_list
```

```python
participate = {"PID":project_list,"UID":user_list}
participate_df = pd.DataFrame(data=participate)
```
